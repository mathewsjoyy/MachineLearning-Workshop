```python
import numpy as np
import matplotlib.pyplot as plt

# Set a random seed for reproducibility
np.random.seed(42)

# Generate 100 random data points within the specified ranges
num_points = 100

x1 = np.random.uniform(-2, 2, num_points)
y1 = np.random.uniform(4, 7, num_points)

x2 = np.random.uniform(3, 7, num_points)
y2 = np.random.uniform(0, 4, num_points)

# Create a scatter plot for the two clusters
plt.scatter(x1, y1, label='Class 1', c='blue')
plt.scatter(x2, y2, label='Class 2', c='red')
plt.grid(True)
plt.legend()
```

```python
# Prompt the user to enter the slope and intercept of the decision boundary
#slope = float(input("Enter the slope of the decision boundary: "))
#intercept = float(input("Enter the intercept of the decision boundary: "))

#slope = 4
#intercept = -1.0


# PROVIDED WITH WRONG NUMBERS
def line(x):
    slope = 4
    intercept = -1
    return slope*x + intercept


## FILL OUT:
# Define a range of x-values for plotting
x_values = np.linspace(-7, 7, 100)
# Calculate the corresponding y-values based on the user-provided parameters
y_values = line(x_values)
plt.plot(x_values, y_values, label='Decision Boundary', color='green')

#x1 = np.random.uniform(-2, 2, num_points)
#y1 = np.random.uniform(4, 7, num_points)

#x2 = np.random.uniform(3, 7, num_points)
#y2 = np.random.uniform(0, 4, num_points)

### PROVIDED:
x = np.concatenate((x1, x2))
y = np.concatenate((y1, y2))

# y = slope*x + intercept --> y - slope*x - intercept < 0

for xi, yi in zip(x, y):
    if yi - line(xi) < 0:
        plt.scatter([xi], [yi], c= 'red')
    else:
        plt.scatter([xi], [yi], c = 'blue')


# Plot the decision boundary

#plt.plot([-6, 6], [-4, 8], label = "This also works", color='green')
#plt.scatter(x1, y1, label='Cluster 1', c='blue')
#plt.scatter(x2, y2, label='Cluster 2', c='red')
plt.xlabel('X')
plt.ylabel('Y')
plt.xlim([-7, 7])
plt.ylim([-5, 9])
plt.title('Linear Decision Boundary')
plt.grid(True)

# Show the plot
plt.legend()
plt.show()
```
