```python
import numpy as np
from sklearn.metrics import mean_squared_error
from math import sqrt

# Set a random seed for reproducibility
np.random.seed(0)

# Generate x values
x = np.linspace(-5, 5, 20)

# Generate y values with some noise
noise = np.random.normal(0, 2, 20)  # Mean=0, Standard Deviation=2
y = x**2 + noise

# Split the data into training (first 12) and test (remaining 8) sets
x_train, y_train = x[:12], y[:12]
x_test, y_test = x[12:], y[12:]

# Generate y values for the function y = -4.55x - 0.35
y_function1 = -4.55 * x - 0.35

# Generate y values for the function y = 1.1927x^2 + 0.4719x + 0.99995
y_function2 = 1.1927 * x**2 + 0.4719 * x + 0.99995

# Generate y values for the function y = 0.1527x^3 + 2.157x^2 + 1.603x + 0.53255
y_function3 = 0.1527*x**3 + 2.157*x**2 + 1.603*x + 0.53255

# Calculate RMSE for the linear model on the training data
rmse_train_linear = sqrt(mean_squared_error(y_train, y_function1[:12]))

# Calculate RMSE for the linear model on the test data
rmse_test_linear = sqrt(mean_squared_error(y_test, y_function1[12:]))

# Calculate RMSE for the quadratic model on the training data
rmse_train_quadratic = sqrt(mean_squared_error(y_train, y_function2[:12]))

# Calculate RMSE for the quadratic model on the test data
rmse_test_quadratic = sqrt(mean_squared_error(y_test, y_function2[12:]))

# Calculate RMSE for the qubic model on the training data
rmse_train_qubic = sqrt(mean_squared_error(y_train, y_function3[:12]))

# Calculate RMSE for the qubic model on the test data
rmse_test_qubic = sqrt(mean_squared_error(y_test, y_function3[12:]))

# Print RMSE for both models and datasets
print("RMSE for Linear Model (Training):", rmse_train_linear)
print("RMSE for Linear Model (Test):", rmse_test_linear)
print("RMSE for Quadratic Model (Training):", rmse_train_quadratic)
print("RMSE for Quadratic Model (Test):", rmse_test_quadratic)
print("RMSE for Qubic Model (Training):", rmse_train_qubic)
print("RMSE for Qubic Model (Test):", rmse_test_qubic)
```

```python
import matplotlib.pyplot as plt

# Create a scatter plot to visualize the training and test data
plt.scatter(x_train, y_train, label='Training Data', color='b')
plt.scatter(x_test, y_test, label='Test Data', color='r')

# Plot the function y = -4.55x - 0.35
plt.plot(x, y_function1, label='y = -4.55x - 0.35', color='g')

# Plot the function y = 1.1927x^2 + 0.4719x + 0.99995
plt.plot(x, y_function2, label='y = 1.1927x^2 + 0.4719x + 0.99995', color='m')

# Plot the function y = 0.1527x^3 + 2.157x^2 + 1.603x + 0.53255
plt.plot(x,y_function3, label='y = 0.1527x^3 + 2.157x^2 + 1.603x + 0.53255', color='k')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Training and Test Data with Functions')
plt.grid(True)

plt.legend()
plt.show()
```
